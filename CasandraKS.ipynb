```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/kick_starter.csv')

# Print the first 10 rows of the DataFrame in a table format
print("First 10 rows of the DataFrame:")
display(df.head(10))

# Print the shape of the DataFrame
print("\nShape of the DataFrame:")
print(df.shape)
```

## Dropped usd_pledged coloumn

```python
df = df.drop(columns=['usd pledged'])

print("DataFrame after dropping 'usd_pledged' column:")
display(df.head())
```

## unifing column names

```python
df = df.rename(columns={'ID': 'id'})

print("DataFrame after renaming 'ID' to 'id':")
display(df.head())
```

## Checking for duplicates

```python
print(f"DataFrame shape before dropping duplicates: {df.shape}")

duplicate_rows = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_rows}")

df = df.drop_duplicates()

print(f"DataFrame shape after dropping duplicates: {df.shape}")
display(df.head())
```

## Handling missing values

```python
print("Missing values per column:")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
print("DataFrame Info (including data types):")
df.info()
```

```python
print(f"DataFrame shape before dropping missing values in 'name': {df.shape}")

df = df.dropna(subset=['name'])

print(f"DataFrame shape after dropping missing values in 'name': {df.shape}")
```

## Drop projects where state = live

```python
print(f"DataFrame shape before dropping 'live' projects: {df.shape}")

df = df[df['state'] != 'live']

print(f"DataFrame shape after dropping 'live' projects: {df.shape}")
```

## CampaignDuration (in days)

```python
df['launched'] = pd.to_datetime(df['launched'])
df['deadline'] = pd.to_datetime(df['deadline'])

df['CampaignDuration'] = (df['deadline'] - df['launched']).dt.days

print("DataFrame with 'CampaignDuration' column:")
display(df.head())
print("\nData type of 'CampaignDuration' column:")
print(df['CampaignDuration'].dtype)
```

## FundingRatio
FundingRatio =
usd_pledged_real/
usd_goal_real

```python
df['FundingRatio'] = df['usd_pledged_real'] / df['usd_goal_real']

print("DataFrame with 'FundingRatio' column:")
display(df.head())
```

## GoalCategory
– Low Goal (< $10,000)
– Medium Goal ($10,000 – $100,000)
– High Goal (> $100,000)


```python
bins = [-float('inf'), 10000, 100000, float('inf')]
labels = ['Low Goal', 'Medium Goal', 'High Goal']

df['GoalCategory'] = pd.cut(df['usd_goal_real'], bins=bins, labels=labels, right=False)

print("DataFrame with 'GoalCategory' column:")
display(df.head())
print("\nValue counts for 'GoalCategory':")
print(df['GoalCategory'].value_counts())
```

##  LaunchMonth


```python
df['LaunchMonth'] = df['launched'].dt.month

print("DataFrame with 'LaunchMonth' column:")
display(df.head())
```

## Create the Table

```python
session.execute("""
DROP TABLE IF EXISTS ks1.kickstarter;
""")
session.execute("""
CREATE TABLE IF NOT EXISTS ks1.kickstarter (
    id int PRIMARY KEY,
    name text,
    category text,
    main_category text,
    currency text,
    deadline date,
    goal double,
    launched timestamp,
    pledged double,
    state text,
    backers int,
    country text,
    usd_pledged_real double,
    usd_goal_real double,
    campaignduration int,
    fundingratio double,
    goalcategory text,
    launchmonth int
)
""")
```

### The deadline and launched columns were converted to datetime using pandas to match the date and timestamp data types defined in the Cassandra table schema

```python
df['deadline'] = pd.to_datetime(df['deadline'])
df['launched'] = pd.to_datetime(df['launched'])
```

## Insert first 5000 rows into table

```python
import datetime

df = df.reset_index(drop=True)  # Reset the index after dropping rows
i, j = df.shape

# Loop through each row and insert into Cassandra, limiting to first 5000 rows
for x in range(0, min(5000, i)):
    session.execute("""
    INSERT INTO ks1.kickstarter (
        id, name, category, main_category, currency,
        deadline, goal, launched, pledged, state,
        backers, country, usd_pledged_real, usd_goal_real,
        campaignduration, fundingratio, goalcategory, launchmonth
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, (
        int(df.loc[x, 'id']),
        df.loc[x, 'name'],
        df.loc[x, 'category'],
        df.loc[x, 'main_category'],
        df.loc[x, 'currency'],
        df.loc[x, 'deadline'].date(),          # Convert to native Python date object
        float(df.loc[x, 'goal']),
        df.loc[x, 'launched'].to_pydatetime(), # Convert to native Python datetime object
        float(df.loc[x, 'pledged']),
        df.loc[x, 'state'],
        int(df.loc[x, 'backers']),
        df.loc[x, 'country'],
        float(df.loc[x, 'usd_pledged_real']),
        float(df.loc[x, 'usd_goal_real']),
        int(df.loc[x, 'CampaignDuration']),
        float(df.loc[x, 'FundingRatio']),
        df.loc[x, 'GoalCategory'],
        int(df.loc[x, 'LaunchMonth'])
    ))
    print(df.loc[x, 'name'], "has been inserted")

# Check inserted rows
rows = session.execute("SELECT * FROM ks1.kickstarter;")
for row in rows:
    print(row)
```

# Queries

## 1) What is the success rate for each main category?

```python
success_rate_by_category = df_cassandra.groupby('main_category')['state'] \
                                       .apply(lambda x: (x == 'successful').sum() / len(x) * 100)

print("Success rate for each main category:")
print(success_rate_by_category.sort_values(ascending=False))
```

## 2) Which countries have the highest average pledged amount in USD?

```python
average_pledged_by_country = df_cassandra.groupby('country')['usd_pledged_real'] \
                                        .mean() \
                                        .sort_values(ascending=False)

print("Countries with the highest average pledged amount in USD:")
print(average_pledged_by_country.head(10))
```

## 3) Which 5 projects have the highest funding ratio?

```python
highest_funding_ratio_projects = df_cassandra.sort_values(by='fundingratio', ascending=False).head(5)

print("Top 5 projects with the highest funding ratio:")
display(highest_funding_ratio_projects[['name', 'fundingratio']])
```

## 4) Among failed projects, which category has the highest number of failures?

```python
failed_projects = df_cassandra[df_cassandra['state'] == 'failed']

# Count number of failed projects per main category
failed_categories_count = failed_projects['main_category'].value_counts()

print("Category with the highest number of failures:")
print(failed_categories_count.head(1))
```

## 5)What is the average campaign duration for successful vs failed projects?

```python
average_duration_by_state = df_cassandra.groupby('state')['campaignduration'].mean()

print("Average campaign duration for successful vs. failed projects:")
print(average_duration_by_state[['successful', 'failed']])
```

## 6)Which month has the highest number of successful projects?

```python
successful_projects = df_cassandra[df_cassandra['state'] == 'successful']

# Count successful projects by launch month
successful_projects_by_month = successful_projects['launchmonth'].value_counts().sort_index()

# Find the month with the highest number of successful projects
highest_month = successful_projects_by_month.idxmax()
max_projects = successful_projects_by_month.max()

print(f"The month with the highest number of successful projects is month {highest_month} with {max_projects} projects.")
print("Successful projects by month:")
print(successful_projects_by_month)
```

## 7)What is the average number of backers per main category?

```python
average_backers_by_category = df_cassandra.groupby('main_category')['backers'].mean().sort_values(ascending=False)
print("Average number of backers per main category:")
print(average_backers_by_category)
```

## 8)What percentage of projects reached or exceeded their goal?

```python
projects_reached_goal = df_cassandra[df_cassandra['usd_pledged_real'] >= df_cassandra['usd_goal_real']]

# Calculate the percentage
percentage_reached_goal = (len(projects_reached_goal) / len(df_cassandra)) * 100

print(f"Percentage of projects that reached or exceeded their goal: {percentage_reached_goal:.2f}%")
```
