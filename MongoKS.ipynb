#**Queries**

## 1)What is the success rate for each main category?

```python
pipeline = [
    {
        "$group": {
            "_id": "$main_category",
            "total_projects": {"$sum": 1},
            "successful_projects": {
                "$sum": {"$cond": [{"$eq": ["$state", "successful"]}, 1, 0]}
            }
        }
    },
    {
        "$project": {
            "success_rate": {
                "$multiply": [
                    {"$divide": ["$successful_projects", "$total_projects"]},
                    100
                ]
            }
        }
    },
    {"$sort": {"success_rate": -1}}
]

results = list(collection.aggregate(pipeline))

print("Success rate for each main category:")

for r in results:
    print(f"{r['_id']:<15} {r['success_rate']:.6f}")
```

## 2)Which countries have the highest average pledged amount in USD?

```python
pipeline = [
    {
        "$group":{
            "_id":"$country",
            "avg_pledged":{"$avg":"$usd_pledged_real"}
        }
    },
    {"$sort":{"avg_pledged":-1}}
]

results=list(collection.aggregate(pipeline))

print("Average pledged amount per country:")
for r in results:
    print(f"{r['_id']:<10} {r['avg_pledged']:.2f}")
```

## 3)Which 5 projects have the highest funding ratio?

```python
pipeline = [
    {"$sort":{"FundingRatio":-1}},
    {"$limit":5},
    {
        "$project":{
            "name":1,
            "FundingRatio":1
        }
    }
]

results=list(collection.aggregate(pipeline))

print("Top 5 projects with highest funding ratio:")
for r in results:
    print(f"{r['name']:<60} {r['FundingRatio']:.4f}")
```

## 4)Among failed projects, which category has the highest number of failures?

```python
pipeline = [
    {"$match": {"state": "failed"}},  # Only failed projects
    {
        "$group": {
            "_id": "$main_category",
            "failures": {"$sum": 1}      # Count failures per category
        }
    },
    {"$sort": {"failures": -1}},        # Sort descending
    {"$limit": 1}                        # Only take the top category
]

results = list(collection.aggregate(pipeline))

print("Category with the highest number of failures:")
for r in results:
    print(f"{r['_id']:<15} {r['failures']}")
```

## 5)What is the average campaign duration for successful vs failed projects?

```python
pipeline = [
    {
        "$match": {"state": {"$in": ["successful", "failed"]}}  # Only include these two states
    },
    {
        "$project": {
            "state": 1,
            "duration": {
                "$divide": [
                    {"$subtract": [{"$toDate": "$deadline"}, {"$toDate": "$launched"}]},
                    1000*60*60*24  # Convert milliseconds to days
                ]
            }
        }
    },
    {
        "$group": {
            "_id": "$state",
            "avg_duration": {"$avg": "$duration"}
        }
    }
]

results = list(collection.aggregate(pipeline))

print("Average campaign duration for successful vs failed projects:")
for r in results:
    print(f"{r['_id']:<12} {r['avg_duration']:.2f} days")
```

## 6)Which month has the highest number of successful projects?

```python
# Pipeline to get number of successful projects per month
pipeline = [
    {"$match": {"state": "successful"}},  # Only successful projects
    {
        "$project": {
            "LaunchMonth": {"$month": {"$toDate": "$launched"}}
        }
    },
    {
        "$group": {
            "_id": "$LaunchMonth",
            "count": {"$sum": 1}
        }
    },
    {"$sort": {"_id": 1}}  # Sort by month ascending
]

results = list(collection.aggregate(pipeline))

# Create a dictionary to simulate Pandas Series output
month_counts = {r["_id"]: r["count"] for r in results}

# Find the month with the highest number of successful projects
max_month = max(month_counts, key=month_counts.get)
max_count = month_counts[max_month]

print(f"The month with the highest number of successful projects is month {max_month} with {max_count} projects.\n")
print("Successful projects by month:")
print("LaunchMonth")
for month in range(1, 13):
    count = month_counts.get(month, 0)
    print(f"{month:<5} {count}")
```

## 7)What is the average number of backers per main category?

```python
pipeline=[
    {
        "$group":{
            "_id":"$main_category",
            "avg_backers":{"$avg":"$backers"}
        }
    },
    {"$sort":{"avg_backers":-1}}
]

results=list(collection.aggregate(pipeline))

print("Average number of backers per category:")
for r in results:
    print(f"{r['_id']:<15} {r['avg_backers']:.2f}")
```

## 8)What percentage of projects reached or exceeded their goal?

```python
pipeline=[
    {
        "$group":{
            "_id":None,
            "total":{"$sum":1},
            "reached":{
                "$sum":{
                    "$cond":[
                        {"$gte":["$usd_pledged_real","$usd_goal_real"]},
                        1,
                        0
                    ]
                }
            }
        }
    },
    {
        "$project":{
            "percentage":{
                "$multiply":[{"$divide":["$reached","$total"]},100]
            }
        }
    }
]

results=list(collection.aggregate(pipeline))

print("Percentage of projects that reached or exceeded their goal:")
for r in results:
    print(f"{r['percentage']:.2f}%")
```
